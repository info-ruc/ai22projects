```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv("BookStacks_all_scores.csv")
data.head()
```

```python
data2 = data.drop(["书名字数完结状态评分链接"],axis=1)
data2.head()
```

```python
length = len(data2)
for i in range(0,length):
    data2.loc[i,'评分2'] = data2.loc[i,'评分'].split('(')[0].split('：')[1]
    print(data2.loc[i,'评分2'])
```

```python
data2.head()
```

```python
data2 = data2.drop("评分",axis=1)
data2.head()
```

```python
data2 = pd.get_dummies(data2, columns=["主标签"])
data2.head()
```

```python
data2 = data2.drop('副标签',axis=1)
data2 = data2.drop('标题分数',axis=1)
data2.head()
```

```python
print(length)
```

```python
data2 = data2.drop('链接',axis=1)
data2.head()
```

```python
for i in range(0,length):
    if data2.loc[i,'完结状态'] == '完结':
        data2.loc[i,'完结状态'] = 1
    elif data2.loc[i,'完结状态'] == '连载':
        data2.loc[i,'完结状态'] = 0
    else:
         data2.loc[i,'完结状态'] = -1
data2.head()
print(type(data2.loc[1,'完结状态']))
print((data2.loc[1,'完结状态']))
print(type('完结'))
data2.head()
```

```python
data2 = data2.drop('评论',axis=1)
data2.head()
```

```python
import numpy as np
data_num = data2['评论分数']
data_num.head()
```

```python
# 将数据转换为浮点数
data_num = data_num.astype(np.float)
# 计算列的均值
mean = np.mean(data_num)
# 计算列的标准差
std = np.std(data_num)
print(std)
# 减去均值并除以标准差
for i in range(0,length):
    data2.loc[i,'评论分数'] = (data2.loc[i,'评论分数'] - mean) / std
data2.head()
```

```python
#去除字数少于一万的
contains_wan = data2['字数'].str.contains("万")
not_contains_wan = ~contains_wan
data2 = data2.drop(data2[not_contains_wan].index)
print(len(data2))
data2.head()
```

```python
#重置索引
data3 = data2.reset_index(drop=True)
data3.loc[96]
```

```python
#删除万字
data3 = data2.reset_index(drop=True)
length_3 = len(data3)
for i in range(0,length_3):
    data3.loc[i,'字数'] = data3.loc[i,'字数'].split('万')[0]
    print(data3.loc[i,'字数'])
data3.head()
```

```python
#对字数进行归一化处理
from sklearn.preprocessing import StandardScaler

# 定义 StandardScaler 对象
scaler = StandardScaler()
# 对字数列进行归一化处理
data3["字数"] = scaler.fit_transform(data3[["字数"]])
data3.head()
```

```python
data3.rename(columns={"评分2":"评分"}, inplace=True)
data3.head()
```

```python
#划分数据集
from sklearn.model_selection import train_test_split
X = data3.drop("评分", axis=1)
X = data3.drop("书名", axis=1)
y = data3["评分"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold

# 加载数据集，这里假设已经将数据集划分成了 X 和 y 两个数组
X2 = X.values
y2 = y.values

# 设置 k 值为 5
k = 5

# 创建 KFold 对象，并设置 k 值
kf = KFold(n_splits=k)

# 定义一个列表用于存储不同最大深度值对应的验证分数
val_scores = []

# 循环遍历不同的最大深度值
for max_depth in range(1, 11):
    # 创建决策树模型
    model = DecisionTreeClassifier(max_depth=max_depth)

    # 使用 KFold 对象划分数据集
    for train_index,val_index in kf.split(X):
        # 获取划分出的训练集和验证集
        X_train, X_val = X2[train_index], X2[val_index]
        y_train, y_val = y2[train_index], y2[val_index]
        # 在这里使用 X_train, y_train 训练模型
        model.fit(X_train, y_train)

        # 在这里使用 X_val, y_val 评估模型
        val_score = model.score(X_val, y_val)

        # 将本次验证分数存储到列表中
        val_scores.append(val_score)

    # 计算本次最大深度对应的平均验证分数
    mean_val_score = np.mean(val_scores)

    # 打印本次最大深度对应的平均验证分数
    print(f"max_depth={max_depth}时，平均验证分数={mean_val_score:.3f}")

```

```python
model = DecisionTreeClassifier(max_depth=10)
model = model.fit(X_train,y_train)
#计算预测值
y_pred = model.predict(X_test)
#均方误差验证
from sklearn.metrics import mean_squared_error
mse_score = mean_squared_error(y_test, y_pred)
print('该回归模型的MSE是：%.2f' %(mse_score))
#计算均方根误差
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print('该回归模型的RMSE是: %.2f' %(mae))
```
